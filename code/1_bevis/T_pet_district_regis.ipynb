```python
import pandas as pd
import os
```

```python
city_dict = {
    "NewTaipei":"新北市",
    "Taipei":"臺北市",
    "Taoyuan":"桃園市",
    "Taichung":"台中市",
    "Tainan":"台南市",
    "Kaohsiung":"高雄市"
}

city_list = list(city_dict.keys())

new_col = ["city", "district", "date", "animal", "regis_count", "removal_count", "update_date"]

df_main = pd.DataFrame(columns=new_col)
```

```python
for city in city_list:
    folder = r"C:\Users\Bevis\Documents\Python_project\sample\output\district_regis"
    file = f"{city}.csv"
    file_path = os.path.join(folder, file)

    df = pd.read_csv(file_path)

    df_edit = df.copy()
    
    # 先還原欄位名
    columns = [
        "area_id",
        "district",
        "登記單位數",
        "regis_count",
        "removal_count",
        "轉讓數",
        "變更數",
        "絕育數",
        "絕育除戶數",
        "免絕育數",
        "免絕育除戶數",
        "animal",
        "date",
        "city",
        "update_date"
    ]

    df_edit.columns = columns

    # 將不要的欄位去除
    df_edit.drop(columns=["area_id", "登記單位數", "轉讓數", "變更數", "絕育數", "絕育除戶數", "免絕育數", "免絕育除戶數"], axis=1, inplace=True)
    
    # 將區的郵遞區號去除，將市的名稱還原
    df_edit["district"] = df_edit["district"].apply(lambda x: x[3:])
    df_edit["city"] = "新北市"

    # 將欄位重新排序
    new_col = ["city", "district", "date", "animal", "regis_count", "removal_count", "update_date"]
    df_edit = df_edit[new_col]
    
    # 合併至主要df
    df_main = pd.concat([df_main, df_edit])
```

```python
# 完成後存檔
main_file = "pet_regis_data.csv"
main_path = os.path.join(folder, main_file)

df_main.to_csv(main_path, index=False, encoding="utf-8")
```
