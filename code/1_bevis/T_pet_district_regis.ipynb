```python
folder = r"C:\Users\Bevis\Documents\Python_project\sample\output\district_regis"
file = "NewTaipei.csv"
file_path = os.path.join(folder, file)

df = pd.read_csv(file_path)

df_edit = df.copy()
```

```python
city_dict = {
    "NewTaipei":"新北市",
    "Taipei":"臺北市",
    "Taoyuan":"桃園市",
    "Taichung":"台中市",
    "Tainan":"台南市",
    "Kaohsiung":"高雄市"
}

city_list = list(city_dict.keys())
```

```python
# 先還原欄位名
columns = [
    "area_id",
    "district",
    "登記單位數",
    "regis_count",
    "removal_count",
    "轉讓數",
    "變更數",
    "絕育數",
    "絕育除戶數",
    "免絕育數",
    "免絕育除戶數",
    "animal",
    "date",
    "city",
    "update_date"
]

df_edit.columns = columns

# 將不要的欄位去除
df_edit.drop(columns=["area_id", "登記單位數", "轉讓數", "變更數", "絕育數", "絕育除戶數", "免絕育數", "免絕育除戶數"], axis=1, inplace=True)

df_edit
```
