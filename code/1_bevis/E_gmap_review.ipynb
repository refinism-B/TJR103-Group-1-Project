```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
import time
```

```python
"""
需要gmap url才能透過selenium去抓
"""

url = r"https://maps.google.com/?cid=4563726407048240698"
chrome_driver_path = r"C:\Users\add41\Documents\Data_Engineer\Project\Flights-Data-Crawler\tool\chromedriver.exe"
service = Service(executable_path=chrome_driver_path)

options = Options()
options.add_argument("--headless")
print('建立連線')

# 根據df_list中的link欄位跑回圈，逐一進入網頁取得html編碼
with webdriver.Chrome(service=service, options=options) as driver:
    driver.get(url)

    # 網頁內有JavaScript動態生成內容，故設定等待網頁讀取完畢後再動作
    wait = WebDriverWait(driver, 10)

    button = driver.find_element(By.XPATH, '//button[contains(@class, "hh2c6") and @data-tab-index="1"]')
    button.click()

    time.sleep(2)

    sidebar = driver.find_element(By.CSS_SELECTOR, '.m6QErb.DxyBCb.kA9KIf.dS8AEf.XiKgde')

    time.sleep(1)

    for _ in range(5):
        driver.execute_script("arguments[0].scrollTop = arguments[0].scrollHeight;", sidebar)
        time.sleep(1)


    page = driver.page_source

soup = BeautifulSoup(page, 'html.parser')

html_str = soup.prettify()

with open('output02.html', 'w', encoding='utf-8') as f:
    f.write(html_str)
```
