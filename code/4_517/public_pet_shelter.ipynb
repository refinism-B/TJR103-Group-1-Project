```python
import pandas as pd
import requests
import os
import certifi
```

# Define the API link

```python
API_LINK = "https://data.moa.gov.tw/Service/OpenData/TransService.aspx?UnitId=2thVboChxuKs"
```

```python
# request needs the SSL authentication but this API doesn't provide it.
# So we set the verify to False, remembering that this behavior is dangerous.
# Do not use it in the unsafe net area.
response = requests.get(API_LINK, verify=False)
data = response.json()
data
```

# Transform to DataFrame and save the original data.

```python
df = pd.DataFrame(data)

# Save the original data
path = "../../data/raw/public_pet_shelter.csv"
os.makedirs(os.path.dirname(path), exist_ok=True)
df.to_csv(path, index=False)
print("Raw data has been saved.")
```

# Do the ELT  
We found that the cityname, url and seq columns is needless.  
Let's drop them off.

```python
df = df.drop(columns=["CityName", "Url", "Seq"])
```

# Sort by ID

```python
df = df.sort_values("ID")
```

# Save the processed file.

```python
path = "../../data/processed/public_pet_shelter_ETL.csv"
os.makedirs(os.path.dirname(path), exist_ok=True)
df.to_csv(path, index=False)
print("Processed data has been saved.")
```
